```python
from bs4 import BeautifulSoup
```

```python
def read_file():
    file = open('consumer_report.txt')
    data = file.read()
    file.close()
    return data
```

```python
soup = BeautifulSoup(read_file(),'lxml')
```

```python
#for product types
attrib = {'class':'products-a-z__results__item'}
product_name = soup.find_all(attrs = attrib)
```

```python
for produc in product_name:
    print(produc.string)
```

```python
#For product_links in HTML

product_links = soup.find_all('a',class_='products-a-z__results__item')
```

```python
for link in product_links:
    print('http://www.consumerreports.org' + link['href'])
```

```python
#By dictionary method
#Another method for navigating tag and accesing product_items
product_items = [div.a.string for div in soup.find_all('div',class_= 'crux-body-copy') ]
```

```python
for product in product_items:
    print(product)
```

```python
#Navigating with loops
product_links = [div.a['href'] for div in soup.find_all('div',class_= 'crux-body-copy') ]
```

```python
#product_hyperlinks
for product in product_links:
    print('http://www.consumerreports.org' + product)
```
